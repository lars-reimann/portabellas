# Plotting with Portabellas

This notebook showcases the plotting functionality in portabellas. All plots are interactive — hover over data points, zoom, and pan directly in the notebook.

```python
from portabellas import Column, Table
```

## Column Plots

Column plots are accessed via the `.plot` property on any `Column`.

```python
temperatures = Column("temperature", [15, 18, 22, 25, 30, 28, 24, 19, 16, 14, 20, 26])
sales = Column("sales", [54, 62, 90, 120, 206, 185, 110, 70, 58, 48, 95, 145])
```

### Box Plot

```python
temperatures.plot.box_plot()
```

### Histogram

```python
temperatures.plot.histogram()
```

### Violin Plot

```python
temperatures.plot.violin_plot()
```

## Table Plots

Table plots are accessed via the `.plot` property on any `Table`.

```python
table = Table(
    {
        "temperature": [15, 18, 22, 25, 30, 28, 24, 19, 16, 14, 20, 26],
        "sales": [54, 62, 90, 120, 206, 185, 110, 70, 58, 48, 95, 145],
        "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"],
    }
)
```

### Grid Plots

Box plots, histograms, and violin plots can be created for all (numeric) columns at once.

```python
table.plot.box_plots()
```

```python
table.plot.histograms()
```

```python
table.plot.violin_plots()
```

### Scatter Plot

```python
table.plot.scatter_plot("temperature", ["sales"])
```

### Line Plot

Line plots automatically aggregate duplicate x-values by computing the mean and an optional confidence interval.

```python
table.plot.line_plot("temperature", ["sales"])
```

### 2D Histogram

```python
table.plot.histogram_2d("temperature", "sales")
```

### Correlation Heatmap

```python
table.plot.correlation_heatmap()
```

## Customization

### Title

All plot methods accept a `title` parameter.

```python
temperatures.plot.box_plot(title="Temperature Distribution")
```

### Theme

All plot methods accept a `theme` parameter:

- `"auto"` (default): adapts to your IDE/browser color scheme
- `"light"`: light background
- `"dark"`: dark background

```python
table.plot.scatter_plot("temperature", ["sales"], theme="dark")
```

## Saving Plots

Plots can be saved to file via the `.write` property.

```python
plot = table.plot.scatter_plot("temperature", ["sales"])

# HTML (always available)
plot.write.html_file("scatter.html")

# PNG, SVG, PDF (require the 'plot' extra: pip install portabellas[plot])
# plot.write.png_file("scatter.png")
# plot.write.svg_file("scatter.svg")
# plot.write.pdf_file("scatter.pdf")
```
